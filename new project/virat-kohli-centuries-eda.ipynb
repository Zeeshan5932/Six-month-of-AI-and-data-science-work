```python
import pandas as pd
import vizard
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
import pandas as pd

# Load the dataset
df = pd.read_csv('/kaggle/input/virat-kholi-centuires/Virat_Kohli_Centuries.csv')

# Define the configuration class
class Config:
    PROBLEM_TYPE = 'classification'  # Man of the Match prediction is a classification problem
    DEPENDENT_VARIABLE = 'Man of the Match'  # Target variable is 'Man of the Match'
   
    # Assuming 'Score', 'Strike Rate', and other numerical fields are continuous variables
    CONTINUOUS_INDEPENDENT_VARIABLES = [
        'Score', 'Strike Rate', 'Position', 'Innings'
    ]
    
    # Assuming 'Type of Match', 'Dismissed', 'Captain', 'Against', 'Venue', 'H/A/N' are categorical variables
    CATEGORICAL_INDEPENDENT_VARIABLES = [
        'Type of Match', 'Dismissed', 'Captain', 'Against', 'Venue', 'H/A/N'
    ]

    TEXT_VARIABLES = []

# Instantiate the config class
config = Config()

# Create a Vizard object with the dataframe and the configuration
viz = vizard.Vizard(df, config)

# # Call the methods on the viz object
# viz.count_missing()
# viz.count_unique()
# viz.categorical_variables()
```
