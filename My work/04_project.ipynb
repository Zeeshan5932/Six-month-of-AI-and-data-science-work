```python
# import libraries 
import pandas as pd
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split, GridSearchCV , cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier,GradientBoostingClassifier
from xgboost import XGBClassifier
```

```python
# load data set of titanic
df = sns.load_dataset('titanic')



# split tha data features and target
x = df[['pclass', 'sex', 'age', 'fare' , 'embarked']]
y = df['survived']

# split tha data to train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create a list of model
models = [('RandomForestClassifier',RandomForestClassifier(random_state=42)),
          ('GradientBoostingClassifier' , GradientBoostingClassifier(random_state=42)),
          ('XGb boost' , XGBClassifier(random_state=42))]

best_model = None
best_accuracy = 0.0

# loop through the models
for name,model in models:
    # create a pipeline of each model
    pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoder', OneHotEncoder(handle_unknown='ignore')),
        ('model', model)
    ])
    
    # perform Cross validation
    score = cross_val_score(pipeline, x_train, y_train, cv=5)

    # get the mean of each model
    mean_Accuracy = score.mean()

    # Fit the pipeline on train data
    pipeline.fit(x_train, y_train)

    # Pred the pipeline on tEST  data
    y_pred = pipeline.predict(x_test)

    # Calculate the accuracy score
    accuracy = accuracy_score(y_test , y_pred)

    # Print the performance metrics
    print('model', name)
    print('cross_validation' , mean_Accuracy)
    print('Test_accuracy' , accuracy)
    print()

    # check if the model is better than the previous model
    if mean_Accuracy > best_accuracy:
        best_accuracy = mean_Accuracy
        best_model = pipeline

print('best_model' , best_model)
```
