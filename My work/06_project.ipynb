### Dealing with missing values 
 - Mean,Median and Mode
 

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load data set of titatnic
df = sns.load_dataset('titanic')
df.head()
```

```python
# Check the missing values
df.isnull().sum().sort_values(ascending=False)
```

```python
# impute missing values with mean
df['age'].fillna(df['age'].mean(), inplace=True)

# Check the missing values
df.isnull().sum().sort_values(ascending=False)
```

```python
# impute missing values with median
df['age'].fillna(df['age'].median(), inplace=True)
```

```python
# impute msiing values with mode
df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

### 02. K-Nearst Neighbour

```python
from sklearn.impute import KNNImputer

# Load the data set of titanic
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
# Call the KNNImputer function with k=3
knn_imputer = KNNImputer(n_neighbors=3)

# Impute the missing values with KNNImputer
df['age'] = knn_imputer.fit_transform(df[['age']])
# Check the missing values
df.isnull().sum().sort_values(ascending=False)
```

# 03.Regression 

```python
# impute missing value with Regression
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Load the data set of titanic
df = sns.load_dataset('titanic')

# Call the IterativeImputer function
reg_imputer = IterativeImputer(max_iter=100, random_state=0)

# Impute the missing values with IterativeImputer
df['age'] = reg_imputer.fit_transform(df[['age']])
```

## 04.Random Forest 

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error, r2_score,mean_absolute_percentage_error
from sklearn.preprocessing import LabelEncoder
```

```python
# Load the data set of titanic
df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

```python
# remove the deck column
df.drop('deck', axis=1, inplace=True)
```

```python
# Columns to encode
columns_to_encode = ['sex', 'embarked', 'who', 'class', 'embark_town', 'alive']

# Dictionary to store LabelEncoders for each column
label_encoders = {}

# Loop to apply LabelEncoder to each column
for col in columns_to_encode:
    # Create a new LabelEncoder for the column
    le = LabelEncoder()

    # Fit and transform the data, then inverse transform it
    df[col] = le.fit_transform(df[col])

    # Store the encoder in the dictionary
    label_encoders[col] = le

# Check the first few rows of the DataFrame
df.head()

```

```python
# Split the dataset into two parts: one with missing values, one without
df_with_missing = df[df['age'].isna()]
# dropna removes all rows with missing values
df_without_missing = df.dropna()

print(df_with_missing.head())
print('\n')
print(df_without_missing.head())
```

```python
print("The shape of the original dataset is: ", df.shape)
print("The shape of the dataset with missing values removed is: ", df_without_missing.shape)
print("The shape of the dataset with missing values is: ", df_with_missing.shape)
```

```python
df_with_missing.head()
```

```python
df_without_missing.head()
```

```python
df.columns
```

```python
# split the data into x and y
X = df_without_missing.drop('age', axis=1)
y = df_without_missing['age']

# split the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.24, random_state=42)

# Random forest imputation
rf_imputer = RandomForestRegressor(n_estimators=100, random_state=42)
rf_imputer.fit(X_train, y_train)


# evalution the metrics
y_pred = rf_imputer.predict(X_test)

print("The RMSE of the model is: ", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2 Score for Random Forest Imputation: ", r2_score(y_test, y_pred))
print("MAE for Random Forest Imputation: ", mean_absolute_error(y_test, y_pred))
print("MAPE for Random Forest Imputation: ", mean_absolute_percentage_error(y_test, y_pred))

```

```python
# check missing values
df.isnull().sum().sort_values(ascending=False)
```

```python
## predict the missing values
y_pred = rf_imputer.predict(df_with_missing.drop('age', axis=1))
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# dn't show the warnings
import warnings
warnings.filterwarnings('ignore')
# replace the missing values with predicted values
df_with_missing['age'] = y_pred

# check the missing values
df_with_missing.isnull().sum().sort_values(ascending=False)
```

```python
# concat the data
df_column = pd.concat([df_with_missing, df_without_missing], axis=0)

# The shape of data frmae
print("The shape of the original dataset is: ", df_column.shape)

df_column.head()
```

```python
for col in columns_to_encode:
    # retrieve the encoder from the dictionary
    le = label_encoders[col]

    # Inverse transform and replace the column in the DataFrame
    df_column[col] = le.inverse_transform(df_column[col])

df_column.head()
```

### 04. Advance technique

```python
# imoprt libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# laod the dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
columns_to_encode =['sex', 'embarked', 'who', 'deck', 'class', 'embark_town', 'alive']

# Store the label encoders in a dictionary so we can use them later
label_encoders = {}

# Loop through the columns to encode
for col in columns_to_encode:
    # create a new label encoder for the column
    le = LabelEncoder()

    # fit & transform the data 
    df[col]= le.fit_transform(df[col])

    # store the encoder into dict
    label_encoders[col] = le

df.head()    
```

```python
# impute missing values with IterativeImputer
reg_imputer = IterativeImputer(max_iter=100, random_state=0)



#impute missing values using IterativeImputer in a for loop for age, embark_town,embarked columns and deck

# Columns to impute
columns_to_impute = ['age', 'embark_town', 'embarked', 'deck']

# Loop to impute each column
for col in columns_to_impute:
    df[col] = reg_imputer.fit_transform(df[[col]]) 
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# Inverse transform the encoded columns
for col in columns_to_encode:
    # retrieve the encoder from the dictionary
    le = label_encoders[col]
    
    # Inverse transform and replace the column in the DataFrame
    df[col] = le.inverse_transform(df[col].astype(int))
    
df.head()
```
