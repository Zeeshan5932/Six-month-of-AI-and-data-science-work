> ### <span style="color:orange;font-size:27px; font-weight: bold;">ML part:</span>
Finally, we can create an initial model for our data. The SimpleClassifier handles all the work for us, implementing the familiar scikit-learn API of fit and predict. Alternatively, we could use the same interface as before and pass the entire dataframe, specifying the target column.

---



```python
ml = dabl.SimpleClassifier(random_state=0)
X = titanic_clean.drop("survived", axis=1)
y = titanic_clean.survived
ml.fit(X, y) 
```


<p style="padding: 15px; background-color: #ffcc99; margin: 0; color: #ff7f00; border: 3px solid #ff7f00; font-family: 'Brush Script MT', cursive;font-weight: bold; font-size: 30px; text-align: center; text-shadow: 1px 1px 3px rgba(0, 0, 0, 0.7); letter-spacing: 1.5px;">(iv) Digits Dataset</p>



```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_digits
X1, y1 = load_digits(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X1, y1, random_state=1)
ml3= dabl.SimpleClassifier().fit(X_train, y_train)
print("Accuracy score", ml3.score(X_test, y_test))
```
