> ***Real-world data can be messy, and dealing with missing values is a common challenge. Luckily, tools like [missingno](https://github.com/ResidentMario/missingno) come to the rescue, offering visualizations to tackle this issue. With over 2k stars on GitHub, this library has gained significant popularity for its effectiveness in handling missing data.***

```python
import pandas as pd
!pip install missingno
import missingno as msno
df = pd.read_csv('/kaggle/input/titanic/train.csv')
```

<p id="2" style="padding: 15px; background-color: black; margin: 0; color: white; border: 3px solid white; font-family: 'Brush Script MT', cursive;font-weight: bold; font-size: 30px; text-align: center; text-shadow: 1px 1px 3px rgba(0, 0, 0, 0.7); letter-spacing: 1.5px;"> 2 |  Missing Values </p>


```python
msno.matrix(df)
```

> ***The abundance of missing values in the Cabin column is apparent upon closer examination. Accompanying this observation is a bar chart on the right, providing a visual summary of data completeness. It highlights the rows with the highest and lowest nullity within the dataset, offering a comprehensive overview of the data's integrity.***msno.matrix(titanic_df)

```python
#with sampling 
msno.matrix(df.sample(50))
```

<p id="3" style="padding: 15px; background-color: black; margin: 0; color: white; border: 3px solid white; font-family: 'Brush Script MT', cursive;font-weight: bold; font-size: 30px; text-align: center; text-shadow: 1px 1px 3px rgba(0, 0, 0, 0.7); letter-spacing: 1.5px;"> 3 | Bar Chart  </p>


> ***We could also opt for a bar chart to visually represent the missing values.***

```python
msno.bar(df)
```

<p id="4" style="padding: 15px; background-color: black; margin: 0; color: white; border: 3px solid white; font-family: 'Brush Script MT', cursive;font-weight: bold; font-size: 30px; text-align: center; text-shadow: 1px 1px 3px rgba(0, 0, 0, 0.7); letter-spacing: 1.5px;"> 4 | HeatMap</p>


> ***`missingno.heatmap` in the `missingno` library calculates nullity correlation, illustrating how the presence or absence of one variable influences another's presence using a heatmap.***

```python
msno.heatmap(df)
```

<p id="5" style="padding: 15px; background-color: black; margin: 0; color: white; border: 3px solid white; font-family: 'Brush Script MT', cursive;font-weight: bold; font-size: 30px; text-align: center; text-shadow: 1px 1px 3px rgba(0, 0, 0, 0.7); letter-spacing: 1.5px;"> 5 |  Dendron </p>


> ***In `missingno`, the dendrogram splits using the minimum distance between clusters created by the hierarchical clustering algorithm.***

```python
msno.dendrogram(df)
```
