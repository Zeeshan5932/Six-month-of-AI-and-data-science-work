```python
#import libraries and modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from colorama import Fore, Style
import plotly.express as px
import plotly.graph_objects as go

# Read all datasets
df1 = pd.read_csv('/kaggle/input/microsft-dataset/msft_stock_data.csv')
df2= pd.read_csv('/kaggle/input/paypal-datsets/pypl_stock_data.csv')
df3=pd.read_csv('/kaggle/input/netflix-datset/netflix.csv')
df4=pd.read_csv('/kaggle/input/adobe-datset/adobe_stock_data.csv')
df5=pd.read_csv('/kaggle/input/google-stock-dataset/google_stock_data.csv')
```

```python
# Text Color
def PrintColor(text: str, color=Fore.GREEN, style=Style.NORMAL):
    "Prints color outputs using colorama using a text F-string"
    print(style + color + text + Style.RESET_ALL)

# Text For Main Heading
def print_yellow_large(text):
    PrintColor(text, Fore.YELLOW + Style.BRIGHT)

# Separator
def print_separator(symbol='-', length=60):
    separator = symbol * length
    PrintColor(separator, Fore.YELLOW + Style.BRIGHT)

# Main Heading
def print_heading(text):
    print_separator()
    PrintColor('╭─────────────────────╮', Fore.YELLOW + Style.BRIGHT)
    print_yellow_large(f"│ {text:^21} │")
    PrintColor('╰─────────────────────╯', Fore.YELLOW + Style.BRIGHT)
    print_separator()

# Function for EDA
def data_overview(tr_d):
    # Display head of the training dataset nicely
    print_heading("The Head Of Dataset")
    print(tr_d.head(5).to_string(index=False))
    print()

    # Shapes of Train
    print_heading("Shape Information")
    PrintColor(f"The Shape Of Train Data: Rows - {tr_d.shape[0]}, Columns - {tr_d.shape[1]}", Fore.MAGENTA)
    print()

    # Info of Train Dataset
    print_heading("Dataset Information")
    PrintColor(f"The Info Of Train Dataset\n{tr_d.info()}", Fore.MAGENTA)
    print()

    # Describe Train
    print_heading("Numerical Summary")
    PrintColor(f"The Numerical Summary of Train\n{tr_d.describe()}", Fore.MAGENTA)
    print()

    # Null Values in Train
    print_heading("Null Values")
    PrintColor("Null Values in Train\n" + str(tr_d.isnull().sum()), Fore.MAGENTA)
    print()

    # Duplicates Values in Train
    print_heading("Duplicate Values")
    PrintColor(f"Duplicates Values in Train: {tr_d.duplicated().sum()}", Fore.MAGENTA)
    print()

    # Visualize numeric features
    numeric_features = tr_d.select_dtypes(include=['float64', 'int64']).columns
    for feature in numeric_features:
        fig = px.histogram(tr_d, x=feature, marginal="box", nbins=20, template="plotly_dark")
        fig.update_layout(
            title=f'Distribution of {feature}',
            xaxis_title=feature,
            yaxis_title='Frequency',
            font=dict(size=14, color='#f0f0f0'),  # Setting font color
            showlegend=False,
            plot_bgcolor='black',  # Setting background color
            paper_bgcolor='black'  # Setting paper background color
        )
        fig.show()

# Example usage:
data_overview(df1)
```

```python
# Read all datasets
microsoft = pd.read_csv('/kaggle/input/microsft-dataset/msft_stock_data.csv')
paypal = pd.read_csv('/kaggle/input/paypal-datsets/pypl_stock_data.csv')
netflix = pd.read_csv('/kaggle/input/netflix-datset/netflix.csv')
adobe = pd.read_csv('/kaggle/input/adobe-datset/adobe_stock_data.csv')
google = pd.read_csv('/kaggle/input/google-stock-dataset/google_stock_data.csv')

# Rename columns to avoid duplicates
microsoft.columns = [col + '_Microsoft' if col != 'Date' else col for col in microsoft.columns]
paypal.columns = [col + '_Paypal' if col != 'Date' else col for col in paypal.columns]
netflix.columns = [col + '_Netflix' if col != 'Date' else col for col in netflix.columns]
adobe.columns = [col + '_Adobe' if col != 'Date' else col for col in adobe.columns]
google.columns = [col + '_Google' if col != 'Date' else col for col in google.columns]

# Merge datasets on 'Date'
merged_data = microsoft.merge(paypal, on='Date', how='outer')
merged_data = merged_data.merge(netflix, on='Date', how='outer')
merged_data = merged_data.merge(adobe, on='Date', how='outer')
merged_data = merged_data.merge(google, on='Date', how='outer')

# Drop any rows with missing values
merged_data.dropna(inplace=True)

# Calculate daily returns for each stock
for col in merged_data.columns[1:]:
    merged_data[col + '_Return'] = merged_data[col].pct_change()


# Define sectors
technology_stocks = ['Microsoft', 'Adobe', 'Google']
fintech_stocks = ['Paypal']
media_stocks = ['Netflix']

# Group by sectors
technology_data = merged_data.filter(like='_Microsoft').join(merged_data.filter(like='_Adobe')).join(merged_data.filter(like='_Google'))
fintech_data = merged_data.filter(like='_Paypal')
media_data = merged_data.filter(like='_Netflix')

# Calculate average daily returns for each sector
technology_returns = technology_data.filter(like='_Return').mean(axis=1)
fintech_returns = fintech_data.filter(like='_Return').mean(axis=1)
media_returns = media_data.filter(like='_Return').mean(axis=1)

# Plot average daily returns for each sector
plt.figure(figsize=(10, 6))
plt.plot(technology_returns.index, technology_returns, label='Technology Sector', color='blue')
plt.plot(fintech_returns.index, fintech_returns, label='Fintech Sector', color='green')
plt.plot(media_returns.index, media_returns, label='Media Sector', color='orange')
plt.xlabel('Date')
plt.ylabel('Average Daily Returns')
plt.title('Average Daily Returns of Different Sectors')
plt.legend()
plt.show()

# Compare volatility across sectors
technology_volatility = technology_data.filter(like='_Return').std().mean()
fintech_volatility = fintech_data.filter(like='_Return').std().mean()
media_volatility = media_data.filter(like='_Return').std().mean()

# Plot volatility comparison
plt.figure(figsize=(8, 5))
plt.bar(['Technology', 'Fintech', 'Media'], [technology_volatility, fintech_volatility, media_volatility], color=['blue', 'green', 'orange'])
plt.xlabel('Sector')
plt.ylabel('Volatility (Standard Deviation)')
plt.title('Volatility Comparison Across Sectors')
plt.show()

# Compare cumulative returns across sectors
technology_cumulative_returns = technology_data.filter(like='_Return').sum(axis=1)
fintech_cumulative_returns = fintech_data.filter(like='_Return').sum(axis=1)
media_cumulative_returns = media_data.filter(like='_Return').sum(axis=1)

# Plot cumulative returns comparison
plt.figure(figsize=(10, 6))
plt.plot(technology_cumulative_returns.index, technology_cumulative_returns, label='Technology Sector', color='blue')
plt.plot(fintech_cumulative_returns.index, fintech_cumulative_returns, label='Fintech Sector', color='green')
plt.plot(media_cumulative_returns.index, media_cumulative_returns, label='Media Sector', color='orange')
plt.xlabel('Date')
plt.ylabel('Cumulative Returns')
plt.title('Cumulative Returns Comparison Across Sectors')
plt.legend()
plt.show()
```
