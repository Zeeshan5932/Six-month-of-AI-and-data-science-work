```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import random as rd
import numpy as np
```

```python
df = pd.read_csv("./clustering.csv")
df.head(5)
```

For this article, we will be taking only two variables from the data – “LoanAmount” and “ApplicantIncome.” This will make it easy to visualize the steps as well. Let’s pick these two variables and visualize the data points:

```python
x = df[["LoanAmount" , "ApplicantIncome"]]
# visulize data point 

plt.scatter(x["ApplicantIncome"] ,x["LoanAmount"] , c='red')
plt.xlabel('AnnualIncome')
plt.ylabel('LoanAmount (In Thousand)')
plt.show()
```

```python
k = 3

centroids = (x.sample(n = k) )

plt.scatter(x["ApplicantIncome"] ,x["LoanAmount"] , c='red')
plt.scatter(centroids['ApplicantIncome'],centroids['LoanAmount'], c='black')
plt.xlabel('AnnualIncome')
plt.ylabel('LoanAmount (In Thousand)')
plt.show()
```

Here, the red dots represent the 3 centroids for each cluster. Note that we have chosen these points randomly, and hence every time you run this code, you might get different centroids.


```python
import warnings
warnings.filterwarnings("ignore")
```

```python
#initial step 1 
# The algorithm iteratively updates the centroids and stops when the centroids don't change
diff = 1
j =0 

while(diff !=0):
    # Step 1 = end 
    
    # step 2:Calculate the Euclidean distance d between each data point and each centroid:
    XD = x
    i = 1
    
    for index1,row_c in centroids.iterrows():
        ED = []
        for index2,row_d in XD.iterrows():
            
            d1 = (row_c['ApplicantIncome'] - row_d['ApplicantIncome']) **2
            d2 = (row_c['LoanAmount'] - row_d['LoanAmount']) ** 2
            
            d = np.sqrt(d1 + d2)
            ED.append(d)
        x[i] = ED    
        i = i+1
       # end step:2
       
       # step_3 : Each point is assigned to the nearest cluster, and this information is stored in the column Cluster. 
    C=[]
    for index,row in x.iterrows():
        min_dist=row[1]
        pos=1
        for i in range(k):
            if row[i+1] < min_dist:
                min_dist = row[i+1]
                pos=i+1
        C.append(pos)
    x["Cluster"]=C
    #---------------------------------------------step:3 end --------------------------------------
    
    # step:4 Update Centroids 
    # Recompute the centroids as the mean of all points in each cluster:
    Centroids_new = x.groupby(["Cluster"]).mean()[["LoanAmount","ApplicantIncome"]]
    # ----------------------- End: 4 ------------------------------------------
    
    # step :5 Check Convergence
    # Calculate the change in centroids:
    if j == 0:
        diff=1
        j=j+1
    else:
        diff = (Centroids_new['LoanAmount'] - Centroids['LoanAmount']).sum() + (Centroids_new['ApplicantIncome'] - Centroids['ApplicantIncome']).sum()
        print(diff.sum())
        # ----------------------- end step : 5
        
        # step: 6 Update the centroids with the new values computed in Step 4.
    Centroids = x.groupby(["Cluster"]).mean()[["LoanAmount","ApplicantIncome"]]
        
            
```

```python
color=['blue','green','cyan']
for k in range(k):
    data=x[x["Cluster"]==k+1]
    plt.scatter(data["ApplicantIncome"],data["LoanAmount"],c=color[k])
plt.scatter(Centroids["ApplicantIncome"],Centroids["LoanAmount"],c='red')
plt.xlabel('Income')
plt.ylabel('Loan Amount (In Thousands)')
plt.show()
```
