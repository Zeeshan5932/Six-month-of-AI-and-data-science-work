```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import seaborn as sns
```

```python
df = pd.read_csv('amazone.csv')
df.head()
```

```python


fig = px.line(df, x='Datetime', y='Close', title='Stock Close Prices Over Time')
fig.show()
```

```python
fig = px.line(df, x='Datetime', y='Open', title='Stock open Prices Over Time')

# Update the layout for background and line colors
fig.update_traces(line=dict(color=' whitesmoke'))  # Set the line color
fig.update_layout(
    plot_bgcolor='rgba(0, 0, 0, 0)',  # Set the plot background color (transparent)
    paper_bgcolor='rgba(0, 0, 0, 0)', # Set the paper background color (transparent)
    font=dict(color='white')           # Set the font color
)

# Update the overall layout for better visibility with a dark theme
fig.update_layout(
    title='Stock Open Prices Over Time',
    xaxis_title='Datetime',
    yaxis_title='Open Price',
    font=dict(
        family="Arial",
        size=12,
        color="white"
    ),
    xaxis=dict(
        showgrid=True,
        gridcolor='rgba(255, 255, 255, 0.2)',
        zerolinecolor='rgba(255, 255, 255, 0.2)'
    ),
    yaxis=dict(
        showgrid=True,
        gridcolor='rgba(255, 255, 255, 0.2)',
        zerolinecolor='rgba(255, 255, 255, 0.2)'
    )
)

fig.show()
```

```python
fig = px.line(df, x='Datetime', y='High', title='Stock high Prices Over Time')

# Update the layout for background and line colors
fig.update_traces(line=dict(color='yellow'))  # Set the line color
fig.update_layout(
    plot_bgcolor='rgba(0, 0, 0, 0)',  # Set the plot background color (transparent)
    paper_bgcolor='rgba(0, 0, 0, 0)', # Set the paper background color (transparent)
    font=dict(color='white')           # Set the font color
)

# Update the overall layout for better visibility with a dark theme
fig.update_layout(
    title='Stock High Prices Over Time',
    xaxis_title='Datetime',
    yaxis_title='Open Price',
    font=dict(
        family="Arial",
        size=12,
        color="white"
    ),
    xaxis=dict(
        showgrid=True,
        gridcolor='rgba(155, 155, 155, 0.2)',
        zerolinecolor='rgba(255, 255, 255, 0.2)'
    ),
    yaxis=dict(
        showgrid=True,
        gridcolor='rgba(255, 255, 255, 0.2)',
        zerolinecolor='rgba(255, 255, 255, 0.2)'
    )
)

fig.show()
```

```python
fig = px.line(df, x='Datetime', y='Low', title='Stock Low Prices Over Time')

# Update the layout for background and line colors
fig.update_traces(line=dict(color='blue'))  # Set the line color
fig.update_layout(
    plot_bgcolor='rgba(0, 0, 0, 0)',  # Set the plot background color (transparent)
    paper_bgcolor='rgba(0, 0, 0, 0)', # Set the paper background color (transparent)
    font=dict(color='white')           # Set the font color
)

# Update the overall layout for better visibility with a dark theme
fig.update_layout(
    title='Stock low Prices Over Time',
    xaxis_title='Datetime',
    yaxis_title='Open Price',
    font=dict(
        family="Arial",
        size=12,
        color="white"
    ),
    xaxis=dict(
        showgrid=True,
        gridcolor='rgba(155, 155, 155, 0.2)',
        zerolinecolor='rgba(255, 255, 255, 0.2)'
    ),
    yaxis=dict(
        showgrid=True,
        gridcolor='rgba(255, 255, 255, 0.2)',
        zerolinecolor='rgba(255, 255, 255, 0.2)'
    )
)

fig.show()
```

```python


fig = make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.2, 
                    subplot_titles=('Candlestick', 'Volume'), row_width=[0.2, 0.7])

# Candlestick
fig.add_trace(go.Candlestick(x=df['Datetime'],
                             open=df['Open'],
                             high=df['High'],
                             low=df['Low'],
                             close=df['Close'],
                             name='Candlestick'), row=1, col=1)

# Volume
fig.add_trace(go.Bar(x=df['Datetime'], y=df['Volume'], name='Volume'), row=2, col=1)

fig.update_layout(title='Stock Data with Volume', xaxis_title='Date', yaxis_title='Price')
fig.show()
```

```python
# Visualization 8: Rolling Mean of Closing Prices
plt.figure(figsize=(12, 6))
df['Close'].rolling(window=30).mean().plot(label='30-Day Rolling Mean', color='red')
df['Close'].plot(label='Close Price', alpha=0.5)
plt.title('30-Day Rolling Mean of Closing Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Distrubution of volume using histogram

plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Volume', bins=20, kde=True)
plt.title('Distribution of Volume')
plt.xlabel('Volume')
plt.ylabel('Count')
plt.show()
```
